**Data set #1 - List of SP500 companies**

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
stock = pd.read_html(url, header=0)[0]
stocks_df = stock['Symbol'].tolist()
stock
```

**Data set #5 - Shares companies**

```python
sharesoutstanding = []

for ticker in stocks_df:
  print(ticker)
  tickerdata = yf.Ticker(ticker)
  try:
    sharesoutstanding.append(tickerdata.info['sharesOutstanding'])
  except:
    sharesoutstanding.append('')

sharesoutstanding_df = pd.DataFrame()
sharesoutstanding_df['sharesoutstanding'] = sharesoutstanding
```

```python
sharesoutstanding_df['index'] = stocks_df
sharesoutstanding_df['sharesOutstanding'] = sharesoutstanding
sharesoutstanding_df.to_excel('/content/drive/MyDrive/DAB103-DEMO/sharesoutstanding_df.xlsx', sheet_name='Sheet', index=False)
```
